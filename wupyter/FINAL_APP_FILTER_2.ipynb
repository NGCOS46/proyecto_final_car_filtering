```python
def apply_filters(driver, make_input, model_input,option_country, option_P, power_min_input, power_max_input, option_gear, option_mileage, option_fuel, option_price):
    # URL y aceptación de cookies
    
    url = 'https://www.autoscout24.com/'
    driver.get(url)
    
    wait = WebDriverWait(driver, 10)
    cookies_accept_button = wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, 'div > div._acceptance-buttons_1i5cd_82 > button._consent-accept_1i5cd_111')))
    cookies_accept_button.click()

    # Navegación a la búsqueda refinada
    refine_search_button = driver.find_element(By.XPATH, '//*[@id="root"]/section[1]/div[1]/div/section/div[2]/div/a[2]')
    refine_search_button = driver.find_element(By.XPATH, '//*[@id="root"]/section[1]/div[1]/div/section/div[2]/div/a[2]')
    driver.execute_script("arguments[0].scrollIntoView(true);", refine_search_button)
    refine_search_button.click()
    time.sleep(5)
    # Aplicar filtros
    make = driver.find_element(By.CSS_SELECTOR, '#make-input-primary-filter')
    make.click()
    make.send_keys(make_input)
    make.send_keys(Keys.RETURN)
    
    time.sleep(5)
    model_input_field = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.CSS_SELECTOR, 'div > div > div.DetailSearchPage_container__O_j_1 > div:nth-child(3) > div:nth-child(1) > div > div.BasicData_container__zw3TT > div:nth-child(2) > div:nth-child(1) > div > div:nth-child(2) > div > div.input-wrapper > input')))
    #model_input_field = driver.find_element(By.CSS_SELECTOR, 'div > div > div.DetailSearchPage_container__O_j_1 > div:nth-child(3) > div:nth-child(1) > div > div.BasicData_container__zw3TT > div:nth-child(2) > div:nth-child(1) > div > div:nth-child(2) > div > div.input-wrapper > input')
    driver.execute_script("arguments[0].value = arguments[1]", model_input_field, model_input)
    model_input_field.send_keys(Keys.RETURN)
    
    time.sleep(5)
    
    country= driver.find_element(By.CSS_SELECTOR, '#country-input > span')
    country.click()
    
    time.sleep(5)
    
    options_country = {
    'Germany': '1',
    'Austria': '2',
    'Belgium': '3',
    'Spain': '4',
    'France': '5',
    'Italy': '6',
    'Luxembourg': '7',
    'Netherlands': '8',
    }
    

    # Get the option index based on the selected country
    option_index = options_country.get(option_country, '1')  # Default to '1' if not found

    time.sleep(2)
    
    # Construct the CSS selector for the selected country option
    country_option_selector = f"#country-input-suggestion-{option_index} > span"
    # Click the selected country option
    country_option = driver.find_element(By.CSS_SELECTOR, country_option_selector)
    country_option.click()
    
    driver.execute_script("window.scrollTo(0, 250)")
    
    time.sleep(5)

    power_choose = driver.find_element(By.CSS_SELECTOR, "div.PowerFilter_powerFilterContainer__GYAHy > div.PowerFilter_powerUnitContainer__RvTXs > div > div.input-wrapper")
    power_choose.click()
    
    time.sleep(5)
    
    HP = driver.find_element(By.CSS_SELECTOR, "#power-unit-select-suggestion-0")
    KW = driver.find_element(By.CSS_SELECTOR, "#power-unit-select-suggestion-1")

    if option_P == 'kw':
        KW.click()
    elif option_P == 'hp':
        HP.click()
    time.sleep(5)
    
    time.sleep(5)
    
    power_min = driver.find_element(By.CSS_SELECTOR, '#power-from')
    power_min.click()
    power_min.send_keys(power_min_input)
    
    time.sleep(5)
    
    
    power_max = driver.find_element(By.CSS_SELECTOR, '#power-to')
    power_max.click()
    power_max.send_keys(power_max_input)

    time.sleep(5)

    gear_choose= driver.find_element(By.CSS_SELECTOR, "#gear-select")
    gear_choose.click()
    
    time.sleep(5)
    
    manual=driver.find_element(By.CSS_SELECTOR, "div > div > div.DetailSearchPage_container__O_j_1 > div:nth-child(3) > div:nth-child(1) > div > fieldset:nth-child(9) > div > div > ul > li:nth-child(1) > label > span > span > span")
    automatic=driver.find_element(By.CSS_SELECTOR, "div > div > div.DetailSearchPage_container__O_j_1 > div:nth-child(3) > div:nth-child(1) > div > fieldset:nth-child(9) > div > div > ul > li:nth-child(2) > label > span > span > span")
    semi=driver.find_element(By.CSS_SELECTOR, "div > div > div.DetailSearchPage_container__O_j_1 > div:nth-child(3) > div:nth-child(1) > div > fieldset:nth-child(9) > div > div > ul > li:nth-child(3) > label > span > span > span")


    if option_gear=='Manual':
        manual.click()
    elif option_gear=='Automatic':
        automatic.click()
    elif option_gear=='Semi':
        semi.click()
        
    time.sleep(5)

    mileage_to = driver.find_element(By.CSS_SELECTOR, "#mileageTo-input")
    mileage_to.click()

    options_mileage = {
        '5.000': 1,
        '10.000': 2,
        '30.000': 4,
        '50.000': 6,
        '80.000': 9,
        '100.000': 11
    }
    
    time.sleep(5)
    
    option_index = options_mileage.get(option_mileage, 1)
    mileage_option = driver.find_element(By.CSS_SELECTOR, f"#mileageTo-input-suggestion-{option_index}")
    mileage_option.click()
    
    time.sleep(5)

    seller_dealer = driver.find_element(By.CSS_SELECTOR, "#seller-type_dealer_radio").click()

    fuel_type = driver.find_element(By.CSS_SELECTOR, "#fuel-type-select")
    fuel_type.click()
    
    time.sleep(5)

    options_fuel = {
        'gasoline': 1,
        'diesel': 2,
        'electric': 4,
        'hybrid': 8
    }
    option_index = options_fuel.get(option_fuel, 1)
    fuel_option = driver.find_element(By.CSS_SELECTOR, f"div.DetailSearchPage_container__O_j_1 > div:nth-child(3) > div:nth-child(1) > div > div:nth-child(3) > div > div > ul > li:nth-child({option_index}) > label > span > span > span.MultiSelect_optionLabel__cBeBI")
    fuel_option.click()
    fuel_type.click()
    
    time.sleep(5)
    
    price_max = driver.find_element(By.CSS_SELECTOR, "#price-to")
    price_max.click()
    
    time.sleep(5)

    options_price = {
        '500': 0,
        '10.000': 12,
        '20.000': 16,
        '30.000': 18,
        '40.000': 19,
        '50.000': 20,
        '75.000': 21,
        '100.000': 22
    }
    option_index = options_price.get(option_price, 0)
    price_option = driver.find_element(By.CSS_SELECTOR, f"#price-to-suggestion-{option_index}")
    price_option.click()
    
    time.sleep(5)

    results_button = driver.find_element(By.CSS_SELECTOR, 'div.DetailSearchPage_container__O_j_1 > div.DetailSearchPage_buttonWrapper__2p0w7 > button')
    results_button.click()
    
    time.sleep(10)
    
    driver.execute_script("window.scrollTo(0, 1050);")
    
    ul_element = driver.find_element(By.CSS_SELECTOR, 'div > div > div.ListPage_wrapper__IxTle > div.ListPage_container__mV15s > main > div.ListPage_pagination__v_4ci > nav > ul')

    li_elements = ul_element.find_elements(By.TAG_NAME, 'li')   
    max_pages=int(len(li_elements) - 4)
    
    
        
   
    driver.execute_script("window.scrollBy(0, -1050);") 
    car_info_list = []  # Create a list to store car information
    page = 0  # Initialize the page number

    while page < max_pages:
        try:
            # Scroll to the next page
            driver.execute_script("window.scrollTo(0, 1050);")

            # Find and click the "Next" button
            next_page_button = driver.find_element(By.XPATH, '//button[@aria-label="Go to page ' + str(page + 2) + '"]')
            next_page_button.click()

            time.sleep(5)
        except:
            break

        
        car_elements = driver.find_elements(By.CLASS_NAME, 'ListItem_wrapper__J_a_C')

        for car_element in car_elements:
    
            make = car_element.find_element(By.CSS_SELECTOR, 'div.ListItem_wrapper__J_a_C > div.ListItem_header__uPzec.ListItem_header_new_design__hPPNh > a > h2').text
            price = car_element.find_element(By.CSS_SELECTOR, 'div.ListItem_wrapper__J_a_C > div.ListItem_listing__VjI4F > div.PriceAndSeals_wrapper__UStkh > p').text
            mileage = car_element.find_element(By.CSS_SELECTOR, 'div.ListItem_wrapper__J_a_C > div.ListItem_listing__VjI4F > div.VehicleDetailTable_container__mUUbY > span:nth-child(1)').text
            gear = car_element.find_element(By.CSS_SELECTOR, 'div.ListItem_wrapper__J_a_C > div.ListItem_listing__VjI4F > div.VehicleDetailTable_container__mUUbY > span:nth-child(2)').text
            fuel = car_element.find_element(By.CSS_SELECTOR, ' div.ListItem_wrapper__J_a_C > div.ListItem_listing__VjI4F > div.VehicleDetailTable_container__mUUbY > span:nth-child(4)').text
            hp = car_element.find_element(By.CSS_SELECTOR, 'div.ListItem_wrapper__J_a_C > div.ListItem_listing__VjI4F > div.VehicleDetailTable_container__mUUbY > span:nth-child(5)').text

            car_info = {
                'make': make,
                'price': price,
                'mileage': mileage,
                'gear': gear,
                'fuel': fuel,
                'hp': hp
            }

            car_info_list.append(car_info)

        page += 1
        
    car_info_frame = pd.DataFrame(car_info_list)
    
    driver.quit()
    
    return car_info_frame
     
    
# Inicialización del driver
chrome_options = Options()
chrome_options.add_argument("--headless")                             #to remove the warning ------ No funciona mirar más adelante
ChromeDriverManager().install()                                       #install
driver = webdriver.Chrome()    
```

```python
def run_apply_filters_for_country(country, filter_criteria):
    make, model, _, option_P, power_min, power_max, option_gear, option_mileage, option_fuel, option_price = filter_criteria
    
    # Inicializa navegador
    chrome_options = Options()
    chrome_options.add_argument("--headless")
    ChromeDriverManager().install()
    driver = webdriver.Chrome()

    # Aplica los filtros para el país actual
    car_info_frame = apply_filters(driver, make, model, country, option_P, power_min, power_max, option_gear, option_mileage, option_fuel, option_price)

    driver.quit()
    return country, car_info_frame



filter_criteria = ("BMW", "3 Series", None, "hp", "100", "200", "Manual", "30.000", "gasoline", "30.000")

# Lista de países disponibles
countries = ["Germany", "Austria", "Belgium", "Spain", "France", "Italy", "Luxembourg", "Netherlands"]

# Ejecuta el filtrado para todos los países en paralelo
with concurrent.futures.ThreadPoolExecutor() as executor:
    results = list(executor.map(lambda country: run_apply_filters_for_country(country, filter_criteria), countries))

# Diccionario para almacenar los DataFrames de resultados
country_data_frames = {}
for country, car_info_frame in results:
    country_data_frames[country] = car_info_frame

# Ahora tienes los DataFrames filtrados para cada país en country_data_frames
for country, df in country_data_frames.items():
    print(f"Resultados para {country}:")
    print(df)
driver.quit()
```

```python
def main():
    st.title("Car Finder App")
    st.sidebar.header("Opciones de Filtro")

    # filtrado de banda strimlit:
    make_input = st.sidebar.text_input("Marca (Make)")
    model_input = st.sidebar.text_input("Modelo (Model)")
    option_P = st.sidebar.radio("Unidad de Potencia", ["hp", "kw"], index=0)
    power_min_input = st.sidebar.number_input("Potencia Mínima", value=100)
    power_max_input = st.sidebar.number_input("Potencia Máxima", value=200)
    option_gear = st.sidebar.radio("Tipo de Transmisión", ["Manual", "Automático", "Semi"], index=0)
    option_mileage = st.sidebar.selectbox("Kilometraje", ["5.000", "10.000", "30.000", "50.000", "80.000", "100.000"], index=2)
    option_fuel = st.sidebar.selectbox("Tipo de Combustible", ["gasolina", "diésel", "eléctrico", "híbrido"], index=0)
    option_price = st.sidebar.selectbox("Precio", ["500", "10.000", "20.000", "30.000", "40.000", "50.000", "75.000", "100.000"], index=3)

    if st.sidebar.button("Aplicar Filtros"):
        # Crea un diccionario por pais
        filter_criteria_by_country = {
            "Germany": (make_input, model_input, "Germany", option_P, power_min_input, power_max_input, option_gear, option_mileage, option_fuel, option_price),
            "Austria": (make_input, model_input, "Austria", option_P, power_min_input, power_max_input, option_gear, option_mileage, option_fuel, option_price),
            "Belgium": (make_input, model_input, "Belgium", option_P, power_min_input, power_max_input, option_gear, option_mileage, option_fuel, option_price),
            "Spain": (make_input, model_input, "Spain", option_P, power_min_input, power_max_input, option_gear, option_mileage, option_fuel, option_price),
            "France": (make_input, model_input, "France", option_P, power_min_input, power_max_input, option_gear, option_mileage, option_fuel, option_price),
            "Italy": (make_input, model_input, "Italy", option_P, power_min_input, power_max_input, option_gear, option_mileage, option_fuel, option_price),
            "Luxembourg": (make_input, model_input, "Luxembourg", option_P, power_min_input, power_max_input, option_gear, option_mileage, option_fuel, option_price),
            "Netherlands": (make_input, model_input, "Netherlands", option_P, power_min_input, power_max_input, option_gear, option_mileage, option_fuel, option_price),
        }

        # Ejecuta en paralelo todos los países
        with concurrent.futures.ThreadPoolExecutor() as executor:
            results = list(executor.map(lambda criteria: run_apply_filters_for_country(*criteria), filter_criteria_by_country.values()))

        country_data_frames = {}
        for country, car_info_frame in results:
            country_data_frames[country] = car_info_frame

        # Muestra los resultados EN tabla en Streamlit
        for country, df in country_data_frames.items():
            st.header(f"Resultados para {country}:")
            st.dataframe(df)
            
if __name__ == "__main__":
    main()
```
