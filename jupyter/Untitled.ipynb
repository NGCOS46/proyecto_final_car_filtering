```python

import pandas as pd
import re
# Specify the file paths for each CSV file
spain_csv_path = 'spain_df.csv'
germany_csv_path = 'germany_df.csv'
italy_csv_path = 'italy_df.csv'
france_csv_path = 'france_df.csv'
belgium_csv_path = 'belgium_df.csv'
netherlands_csv_path = 'netherlands_df.csv'
luxembourg_csv_path = 'luxembourg_df.csv'
austria_csv_path = 'austria_df.csv'

# Read each CSV file into a new DataFrame
spain_df = pd.read_csv(spain_csv_path)
germany_df = pd.read_csv(germany_csv_path)
italy_df = pd.read_csv(italy_csv_path)
france_df = pd.read_csv(france_csv_path)
belgium_df = pd.read_csv(belgium_csv_path)
netherlands_df = pd.read_csv(netherlands_csv_path)
luxembourg_df = pd.read_csv(luxembourg_csv_path)
austria_df = pd.read_csv(austria_csv_path)
```

```python
germany_df['hp'] = germany_df['hp'].str.extract(r'(\d+) hp')
germany_df
```

```python

germany_df['price'] = germany_df['price'].str.replace(r'\.-', '', regex=True)
germany_df
```

```python
germany_df['model'] = germany_df['make'].str.split('\n').str[1]
germany_df
```

```python
germany_df['model'] = germany_df['model'].str.replace(r'(\d+i).*', r'118i', regex=True)
germany_df
```

```python
germany_df = germany_df[germany_df['model'].str.contains('118i', case=False, regex=True)]

germany_df 
```

```python
germany_df['make'] = germany_df['make'].str.extract(r'(BMW)', expand=False)
germany_df
```
