```python
def convert_country_dataframes_to_individual(dataframes_dict):
    individual_dataframes = {}
    
    for country, df in dataframes_dict.items():
        # Genera un nombre de variable válido a partir del nombre del país
        var_name = country.lower().replace(' ', '_') + '_df'
        
        # Agrega el DataFrame a las variables individuales
        individual_dataframes[var_name] = df
    
    return individual_dataframes
```
